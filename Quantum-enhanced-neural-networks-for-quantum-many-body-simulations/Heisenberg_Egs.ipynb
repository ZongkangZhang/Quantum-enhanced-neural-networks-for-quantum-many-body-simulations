```python
from package_hybrid_ansatz import *

try:
    results = torch.load('data/Heisenberg_Egs.pt', weights_only=False)
except FileNotFoundError:
    results = {}

start_nq = 2
end_nq = 16

for Nq in range(start_nq, end_nq + 1):
    if Nq not in results: 

        start_time = datetime.now()
        Eg = calcHeisenbergEg(Nq)
        end_time = datetime.now()
        print(f"elapsed time: ", (end_time - start_time).total_seconds(), "seconds")
        
        # update results
        results[Nq] = Eg
        torch.save(results, 'data/Heisenberg_Egs.pt')
        
        print(f'Nq={Nq}, Eg={Eg} (saved)')
    else:
        print(f'Nq={Nq} already calculated, Eg={results[Nq]}')

print("Done!")
```
